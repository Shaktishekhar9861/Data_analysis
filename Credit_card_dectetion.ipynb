# separeting the data for analysis

```python
legit=card[card.Class==0]
fraud=card[card.Class==1]
print(legit.shape)
```

# compair the values for both traction

```python
legit_sample=legit.sample(n=492)
new_data_set=pd.concat((legit_sample,fraud),axis=0)
new_data_set.head()
```

```python
new_data_set["Class"].value_counts()
```

# Spliting the data into feature and traget

```python
x=new_data_set.drop(columns="Class",axis=1)
y=new_data_set["Class"]
```

### Here x is the feture data set and y is the target data set

```python
print(x)
```

```python
print(y)
```

# split the data into training data and testing data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
print(x.shape,x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# Model Training

```python
reg=LogisticRegression()
```

# Training the logictic regression model with training data

```python
reg.fit(x_train,y_train)
```

## Model Evaluation accuracy score

```python
x_train_predic=reg.predict(x_train)
accuracy=accuracy_score(x_train_predic,y_train)
print(accuracy)
```

```python
x_test_predic=reg.predict(x_test)
accuracy=accuracy_score(x_test_predic,y_test)
print(accuracy)
```
